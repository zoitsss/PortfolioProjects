```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns
import datetime as dt
import re

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)


# reading the data

df = pd.read_csv(r"C:\Users\CC\Documents/movies.csv")
```

```python
df.head()
```

```python
# Let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# looking as data types
df.dtypes
```

```python
# Changing data types to more relevant ones

df['budget'] = df['budget'].astype('float64')
df['gross'] = df['gross'].astype('float64')
```

```python
# Creating new correct year column by taking first four characters from the released col

df['year_correct'] = df['released'].astype(str).str.extract(r'\b(\d{4})\b')
df
```

```python
df = df.sort_values(by=['gross'], inplace = False, ascending = False)             
```

```python
# Looking at distinct companies

df['company'].drop_duplicates().sort_values(ascending=False)
```

#  Finding Correlations

```python
# Scatter Plot Of Budget vs Gross


plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for film')
plt.ylabel('Gross Earning')

plt.show()

```

```python
# Plotting a regression plot b/w budget and gross

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"orange"}, line_kws={"color":"green"})
```

```python
# Plot above shows positive corr between budget and gross

df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matric For Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()


```

```python
# Looking for corr with non numeric values by changing dtype to category and assigning cat codes

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
        
df_numerized        
        
```

```python
# Plotting a corr heat map for above data 

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matric For Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Simplisize for better understanding

corr_mat = df_numerized.corr()

corr_pairs = corr_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
